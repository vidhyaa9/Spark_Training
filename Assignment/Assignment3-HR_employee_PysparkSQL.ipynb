a)Create HR_Employee Pysaprk dataFrame.

```
HR_employee_df = spark.read.csv("dbfs:/FileStore/tables/HR_Employee.csv", header = True, inferSchema =True)
```

b)Show printSchema & show 10 rows of Dataset

```
HR_employee_df.printSchema()
HR_employee_df.show(10, truncate = False)
```

```
from pyspark.sql.functions import *
HR_employee_df.createTempView("HR_employee_data")
```

c)Most employee is working in which department

```
spark.sql("select Department, count(*) as Count_Emp_depart from HR_employee_data group by Department order by  Count_Emp_depart desc").show(1)
```

d)Highest number of job roles

```
spark.sql("select JobRole, count(*) as Count_Jobrole from HR_employee_data group by JobRole order by  Count_Jobrole desc").show(5)
```

e)Which gender have higher strength as workforce?

```
spark.sql("select Gender, count(*) as strength from HR_employee_data group by Gender order by strength desc").show()
```

f)Compare the marital status of employee and find the most frequent status.

```
spark.sql("select MaritalStatus, count(*) as totalCount from HR_employee_data group by MaritalStatus order by totalCount desc").show()
```

g)Mostly hired employee have qualification

```
spark.sql("select Education, count(*) as most_hired_count from HR_employee_data group by Education order by most_hired_count desc").show(1)
```

h)Find the count of employee from which education fields

```
spark.sql("select EducationField, count(*) as total_count from HR_employee_data group by EducationField order by total_count desc").show()
```

i)What is the job satisfaction level of employee?

```
spark.sql("select JobSatisfaction, count(*) as JobSatisfaction_level from HR_employee_data group by JobSatisfaction order by JobSatisfaction_level desc").show()
```

j)Does most of employee do overtime: Yes or No?

```
spark.sql("select OverTime, count(*) as count_overtime from HR_employee_data group by OverTime order by count_overtime desc").show(1)
```

k)Find Min & Max Salaried employees.

```
spark.sql("select max(Income) as max_income, min(Income) as min_income from HR_employee_data").show()
```

l)Does most of the employee do business travel? Find of the employees counts for each category

```
spark.sql("select BusinessTravel, count(*) as BusinessTravel_count from HR_employee_data group by BusinessTravel order by BusinessTravel_count desc").show()
```

n)Find the employee qualification receiving salary lower than equal to avg. salary of all employee.

```
spark.sql("select Education,  count(*) as totalcount from HR_employee_data where Income<=(select AVG(Income) from HR_employee_data) group by Education order by totalcount desc").show()
```
